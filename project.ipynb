### Data preprocessing

```python
temp = movies["genres"][0]
temp
```

```python
import json
```

```python
def helper(genre_string):
  temp = json.loads(genre_string)
  temp_genres = []
  for genre in temp:
    temp_genres.append(genre["name"])
  genre_string = temp_genres
  return genre_string
```

```python
movies["genres"] = movies["genres"].apply(helper)
```

```python
movies.head()
```

```python
movies["keywords"] = movies["keywords"].apply(helper)
```

```python
movies.head()
```

```python
movies["cast"][0]
```

```python
def cast_helper(cast_string):
  temp = json.loads(cast_string)
  temp_cast = []
  for cast in temp[:3]:
    temp_cast.append(cast["name"])
  genre_string = temp_cast
  return genre_string
```

```python
movies["cast"] = movies["cast"].apply(cast_helper)
```

```python
movies.head()
```

```python
movies["crew"][0]
```

```python
def crew_helper(crew_string):
  temp = json.loads(crew_string)
  temp_crew = []
  for crew in temp:
    if (crew["job"] == "Director"):
      temp_crew.append(crew["name"])
  genre_string = temp_crew
  return genre_string
```

```python
movies["crew"] = movies["crew"].apply(crew_helper)
```

```python
movies.head()
```

```python
def overview_helper(overview_string):
  temp = overview_string.split()
  return temp 
```

```python
movies["overview"] = movies["overview"].apply(overview_helper)
```

```python
movies.head()
```

### Remove spaces

```python
movies["genres"] = movies["genres"].apply(lambda x : [i.replace(" ", "") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x : [i.replace(" ", "") for i in x])
movies["cast"] = movies["cast"].apply(lambda x : [i.replace(" ", "") for i in x])
movies["crew"] = movies["crew"].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

```python
df = movies[["movie_id", "title", "tags"]]
```

```python
df.head()
```

### Converting tags list to string

```python
df["tags"] = df["tags"].apply(lambda x : " ".join(x))
```

```python
df.head()
```

### Converting tags to lowercase

```python
df["tags"] = df["tags"].apply(lambda x : x.lower())
```

```python
df.head()
```

### Stemming the tags

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
def fun(tag):
  temp = []
  for word in tag.split():
    temp.append(stemmer.stem(word))
  return " ".join(temp)
```

```python
df["tags"] = df["tags"].apply(fun)
```

```python
df.head()
```

### Vectorizing the tags

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=5000, stop_words="english") 
```

```python
vectors = vectorizer.fit_transform(df["tags"]).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
vectorizer.get_feature_names()
```

### Calculating cosine similarity between vectors

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
# eg - cosine similarity of first movie with all movies
similarity[0]
```

### Build the recommendation system

```python
def recommend(movie_title):
  # getting the index of the movie in the dataframe
  movie_index = df[df["title"] == movie_title].index[0]

  # getting the distance vector for the movie
  distances = similarity[movie_index]

  # enumerating the distances so that the indices are not lost
  # sorting the list according to the distances and keeping the top 5 movies
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x : x[1])[1:6]

  # printing the list of the top 5 similar movies
  for movie in movies_list:
    print(df.iloc[movie[0]]["title"])
```

```python
recommend("Avatar")
```
